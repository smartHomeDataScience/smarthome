```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

import matplotlib.patches as patches
from PIL import Image
import math
```

```python
# First set up the figure, the axis, and the plot element we want to animate
im = np.array(Image.open('living_room.png'), dtype=np.uint8)

x = [280, 286, 297,296]
y = [177, 109, 177,104]
yaw = [0.0, 0.0 , 0.0 ,0.0]
# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

patch = patches.Rectangle((0, 0), 0, 0, fc='y')
```

```python
# initialization function: plot the background of each frame

def init():
    ax.add_patch(patch)
    return patch,
```

```python
# animation function. This is called sequentially
import random
def animate(i):
    center_2d_x = 280
    center_2d_y = 177

    bb_2d_br_x = 307
    bb_2d_br_y = 169

    bb_2d_tl_x = 266
    bb_2d_tl_y = 72


    height =  bb_2d_br_y - bb_2d_tl_y

    weidth =  bb_2d_br_x - bb_2d_tl_x

    patch.set_width(weidth)
    patch.set_height(height)
    patch.set_xy([x[i] , x[i]])
    patch._angle = -np.rad2deg(yaw[i])
    return patch,
    # Create a Rectangle patch
    #rect = patches.Rectangle((center_2d_x ,center_2d_y),weidth,height,linewidth=1,edgecolor='r',facecolor='none')
```

```python
anim = animation.FuncAnimation(fig, animate,
                               init_func=init,
                               frames=len(x),
                               interval=500,
                               blit=True)
plt.show()
HTML(anim.to_html5_video())
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
import math
im = np.array(Image.open('living_room.png'), dtype=np.uint8)

# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

center_2d_x = 280
center_2d_y = 177

bb_2d_br_x = 307
bb_2d_br_y = 169
 
bb_2d_tl_x = 266
bb_2d_tl_y = 72


height =  bb_2d_br_y - bb_2d_tl_y

weidth =  bb_2d_br_x - bb_2d_tl_x

# Create a Rectangle patch
rect = patches.Rectangle((center_2d_x,center_2d_y),weidth,height,linewidth=1,edgecolor='r',facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
